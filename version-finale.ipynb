```python
import tkinter as tk
from tkinter import ttk
# coefficients
coef_ue1 = {
    'R101': 10, 'R102': 10, 'R103': 7, 'R104': 7, 'R106': 5,
    'R107': 0, 'R108': 6, 'R109': 0, 'R110': 5, 'R111': 4, 'R112': 2, 'R113': 5,
    'R114': 5, 'R115': 0, 'SAE11': 20, 'SAE12': 20, 'SAE16': 7
}

coef_ue2 = {
    'R101': 4, 'R102': 0, 'R103': 2, 'R104': 8, 'R105': 6, 'R106':0, 
    'R107':0, 'R108':0, 'R109': 0, 'R110': 5, 'R111': 5, 'R112': 2,
    'R113': 9, 'R114': 9, 'R115': 3, 'SAE11': 0, 'SAE12':0,
    'SAE13': 29,  'SAE14': 0, 'SAE15':0, 'SAE16': 7
}

coef_ue3 = {
    'R101': 4, 'R103': 2, 'R106': 5, 'R107': 15, 'R108': 6,
    'R109': 4, 'R110': 5, 'R111': 5, 'R112': 2, 'R115': 3,
    'SAE14': 20, 'SAE15': 20, 'SAE16': 7
}


all_matieres = sorted(set(coef_ue1) | set(coef_ue2) | set(coef_ue3))

# fenetre

fenetre = tk.Tk()
fenetre.title("Calcul de Moyenne")
fenetre.geometry("400x800")

frame = ttk.Frame(fenetre, padding=10)
frame.pack(fill='both', expand=True)

# Dictionnaire pour stocker les zones de saisie des notes
entry_notes = {}

for matiere in all_matieres:
    row = ttk.Frame(frame)
    row.pack(fill='x', pady=2)

    label = ttk.Label(row, text=matiere, width=10)
    label.pack(side='left')

    entry = ttk.Entry(row, width=10)
    entry.pack(side='left')

    entry_notes[matiere] = entry

# --- Fonction pour calculer les moyennes ---

def calculer_moyenne(coefs, notes):
    total_points = 0
    total_coef = 0
    for matiere, coef in coefs.items():
        try:
            note = float(notes.get(matiere, 0))
        except ValueError:
            note = 0
        total_points += note * coef
        total_coef += coef
    if total_coef == 0:
        return 0
    return round(total_points / total_coef, 2)

# Zone d’affichage des résultats
resultats_label = ttk.Label(frame, text="",)
resultats_label.pack(pady=20)

# Bouton de calcul 
def bouton_action():
   
    notes_utilisateur = {}
    for matiere, entry in entry_notes.items():
        try:
            notes_utilisateur[matiere] = float(entry.get())
        except ValueError:
            notes_utilisateur[matiere] = 0

    # Calcul des moyennes
    moy_ue1 = calculer_moyenne(coef_ue1, notes_utilisateur)
    moy_ue2 = calculer_moyenne(coef_ue2, notes_utilisateur)
    moy_ue3 = calculer_moyenne(coef_ue3, notes_utilisateur)

    # Affichage
    resultats_label.config(text=f"Moyenne UE1 : {moy_ue1}\nMoyenne UE2 : {moy_ue2}\nMoyenne UE3 : {moy_ue3}")

ttk.Button(frame, text="Calculer les moyennes", command=bouton_action).pack(pady=10)

fenetre.mainloop()
```
